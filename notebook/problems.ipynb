# Descriptive Statistics Problems

## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title
df["title_length"] = df["title"].apply(len)

print(df.head())
```

```python
# Calculate mean, median, and mode of the 'title_length' column

from scipy.stats import mode

mean_length = df["title_length"].mean()
print("Mean:", mean_length)

median_length = df["title_length"].median()
print("Median:", median_length)

mode_length = (df["title_length"]).mode()
print("Mode:", mode_length)
```

```python
mode_length[0]
```

```python
median_length
```

```python
# Calculate the range, variance, and standard deviation of 'title_length'
range_length = np.ptp(df["title_length"])
print("Range:", range_length)

variance_length = np.var(df["title_length"])
print("Variance:", variance_length)

std_dev_length = np.std(df["title_length"])
print("Standard Deviation:", std_dev_length)
```

```python
# Calculate skewness and kurtosis of the 'title_length' column

skewness_length = skew(df["title_length"])
print("Skewness:", skewness_length)

kurtosis_length = kurtosis(df["title_length"])
print("Kurtosis:", kurtosis_length)
```

```python
# Print the shortest and longest title based on their length

longest_title = df.loc[df["title_length"].idxmax(), "title"]
shortest_title = df.loc[df["title_length"].idxmin(), "title"]

print(f"The longest movie title is: '{longest_title}'")
print(f"The shortest movie title is: '{shortest_title}'")
```

```python
# plot your results

plt.figure(figsize=(10, 5))

plt.hist(df["title_length"], bins=20, alpha=0.7, color="blue", edgecolor="black", density=True)

plt.title("Distribution of Movie Title Lengths")
plt.xlabel("Title Length (Characters)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))


plt.scatter(df.index, df["title_length"], color="blue", alpha=0.7)

plt.title("Scatterplot of Movie Title Lengths")
plt.xlabel("Movie Index")
plt.ylabel("Title Length (Characters)")
plt.grid(True)
plt.show()
```

In this exercise, you will manually calculate the standard deviation by selecting five actual titles from the movie dataset you loaded earlier.


```python
# Create a list with 5 actual values from df['title_length'], for example: [10, 13, 14, 18, 22]

title_lengths = [10, 13, 14, 18, 22]

```

```python
# Calculate the mean of the 5 values
title_lengths = [10, 13, 14, 18, 22] 

mean_length = sum(title_lengths) / len(title_lengths)
print("Mean:", mean_length)
```

```python
# Calculate the squared differences with respect to the mean

mean_length = sum([10, 13, 14, 18, 22]) / len([10, 13, 14, 18, 22])
print(f"Mean Title Length: {mean_length}")

df_sample = pd.DataFrame({"title_length": [10, 13, 14, 18, 22]})
df_sample["squared_difference"] = (df_sample["title_length"] - mean_length) ** 2

print(df_sample)
```

```python
# Calculate the variance (average of the squared differences) and its square root to obtain the standard deviation

variance_length = sum([29.16, 5.76, 1.96, 6.76, 43.56]) / 5
print(f"Variance: {variance_length}")

std_dev_length = variance_length ** 0.5
print(f"Standard Deviation: {std_dev_length}")
```
